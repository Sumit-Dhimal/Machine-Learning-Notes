# Univariate Analysis

### Univariate analysis is the simplest form of analysis, focusing on describing, summarizing, and identifying patterns within a single variable at a time
### It uses central tendency (mean, median, mode) and dispersion (range, variance, standard deviation) to analyze data without examining relationships or causes

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../05-Understanding-data/train.csv')
```

```python
df.head()
```

# 1. Categorical data

## a. Count plot

```python
sns.countplot(data=df, x='Sex')
#df['Sex'].value_counts().plot(kind='bar')
```

## b. Piechart

```python
# autopct stands for auto percentage.

# It tells matplotlib to automatically display the percentage values on the pie chart.

# %.2f is string formatting meaning 2-values after point
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f') 
```

# 2. Numerical data

## a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'], bins=5)
```

## b. Distplot / Histplot

```python

# KDE = Kernel Density Estimation

# It is a smooth curve that estimates the probability distribution of your data.

# So instead of just bars (histogram), you also get a smooth line showing the data distribution shape.
sns.histplot(df['Age'], kde=True)
```

## c. Boxplot

```python
sns.boxplot(df['Age'])
```
